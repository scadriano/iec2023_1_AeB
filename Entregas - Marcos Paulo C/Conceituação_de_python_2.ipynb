```python
import  pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


print('Bibliotecas importadas com sucesso')

df_dataset = pd.read_csv('https://raw.githubusercontent.com/scadriano/iec2023_1_AeB/main/01.%20Conceitua%C3%A7%C3%A3o%20de%20Python/hospital.csv', sep=',',index_col=None)
print('Dados importados com sucesso')
```

```python
#exibir as primeiras partes do dataframe
display(df_dataset.head())
```

```python
#mostrar todo o dataframe
display(df_dataset)
```

```python
#remover colunas de Estado e Time:
df_dataset = df_dataset.drop(columns=['Estado','Time'])
df_dataset
```

```python
#Fazer um gráfico:
df_dataset.groupby('Diagnóstico').count().plot.pie(y = 'Internação')
```

```python
#Fazendo um  gráfico no Matplotlib
plt.plot(df_dataset['Nome'], df_dataset['Idade'])
```

```python
#Fazendo um  gráfico no Seaborn
fig = sns.scatterplot(data=df_dataset,
                      x='Manchas',
                      hue='Sexo',
                      style='Sexo',
                      y='Nome'
                      )
```
