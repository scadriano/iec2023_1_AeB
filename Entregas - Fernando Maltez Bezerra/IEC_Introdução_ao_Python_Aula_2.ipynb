```python
import matplotlib.pyplot as plt
```

```python
df_fernando = pd.read_csv('https://raw.githubusercontent.com/scadriano/iec2023_1_AeB/main/01.%20Conceitua%C3%A7%C3%A3o%20de%20Python/hospital.csv')


plt.scatter(df_fernando['Nome'], df_fernando['Idade'], marker='d', label='Idade', color='pink')
plt.scatter(df_fernando['Nome'], df_fernando['Peso'], marker='o', label='Peso', color='green')


plt.xlabel('Nome')
plt.ylabel('Valores')

plt.title('Idade e Peso')

plt.legend()

plt.show()
```

```python
df_fernando = pd.read_csv('https://raw.githubusercontent.com/scadriano/iec2023_1_AeB/main/01.%20Conceitua%C3%A7%C3%A3o%20de%20Python/hospital.csv')


sns.set(style='whitegrid')

sns.lineplot(data=df_fernando, x='Idade', y='Temperatura', color='red')

plt.xlabel('Idade', fontsize=16)
plt.ylabel('Temperatura', fontsize=13)

plt.grid(False)

plt.show()
```
