```python
df_dataset.groupby('Diagnóstico').count().plot.pie(y = 'Internação')
```

#Tarefas:

### 1.Construa um gáfico usando a biblioteca Matplotlib

```python
#utilizando a Matplotlib
from matplotlib.pyplot import pie, axis, show

sums = df_dataset.groupby(df_dataset["Diagnóstico"])["Internação"].sum()
axis('equal');
pie(sums, labels=sums.index);
show()
```

### 2.Construa um gáfico usando a biblioteca Seaborn
