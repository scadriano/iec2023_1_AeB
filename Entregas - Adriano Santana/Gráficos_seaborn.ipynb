## Importação das bibliotecas

```python
import seaborn as sns
import pandas as pd
```

## Carregamento da base de dados

```python
listaDados = pd.read_csv('/content/plano_saude (1).csv', encoding='unicode_escape')
```

```python
display(listaDados)
```

```python
sns.relplot(x="custo", y="Idade", data=listaDados);
```

```python
sns.relplot(x="custo", y="Idade", hue="genero", data=listaDados);
```

```python
sns.relplot(x="custo", y="Idade", hue="genero", style="genero", data=listaDados);
```

```python
sns.relplot(x="custo", y="Idade", hue="genero", style="fumante", data=listaDados);
```

```python
sns.relplot(x="custo", y="Idade", hue="tamanho", data=listaDados);
```

```python
sns.relplot(x="custo", y="Idade", size="genero", data=listaDados);
```

```python
sns.relplot(x="custo", y="Idade", size="tamanho", sizes=(15, 200), data=listaDados);
```
