```python
df_dataset = pd.read_csv('https://raw.githubusercontent.com/lucasiqueira/iec2023_1_AeB/lucas-siqueira-iec/01.%20Conceitua%C3%A7%C3%A3o%20de%20Python/hospital.csv', sep=',', index_col=None)
```

```python
df_dataset = df_dataset.drop(columns=['Estado', 'Time'])
display(df_dataset)
```

```python
df_dataset.groupby('Diagnóstico').count().plot.pie(y = 'Internação')
```

Tarefa 1: Construa um gráfico usando a biblioteca Matplotlib

Tarefa 2: Construa um gráfico usando a biblioteca Seaborn

```python
import matplotlib.pyplot as plt

df_dataset['Tipo de Febre'] = df_dataset['Temperatura'].apply(lambda x: 'Febre baixa' if x < 39 else 'Febre alta')

df_febre_grouped = df_dataset.groupby('Tipo de Febre').size()

plt.pie(df_febre_grouped, labels= df_febre_grouped.index, autopct='%1.1f%%')
plt.title('Distribuição dos Tipos de Febre')
plt.show()
```

```python
import seaborn as sns

sns.set(
    style='darkgrid',
    palette='Set1',
    font_scale=1.2,
    rc={'figure.figsize': (8, 6)}
)

plt.figure(figsize=(6, 6))
sns.countplot(x='Tipo de Febre', data=df_dataset)
plt.ylabel('Quantidade de Pacientes')
plt.title('Distribuição dos Tipos de Febre')

plt.show()
```
