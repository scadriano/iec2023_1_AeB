```python
# importar as bibliotecas - Visualizar dados!
import matplotlib.pyplot as plt
import seaborn as ans
import pandas as pd
print('Biblioteca carregada com sucesso!')
```

```python
df_dataset = pd.read_csv ('https://raw.githubusercontent.com/FredericoMiguel/iec2023_1_AeB/main/01.%20Conceitua%C3%A7%C3%A3o%20de%20Python/hospital.csv', sep=',', index_col=None)
print('Dados importados com sucesso!')
```

```python
# exibir uma amostra do dataframe
display(df_dataset.head())
```

```python
# exibir todo o dataframe
display(df_dataset)
```

```python
# remover as colunas id_planta e cidade_origem
df_dataset = df_dataset.drop(columns=['Estado','Time'])
```

```python
# imprimir o dataframe
display(df_dataset)
```

```python
df_dataset.groupby('Diagnóstico').count().plot.pie(y = 'Internação')
```

```python
x = df_dataset['Idade']
y = df_dataset['Nome']
plt.plot(x, y)
plt.show()
```

```python
ans.scatterplot(x = 'Idade', y = 'Time', data = df_dataset)
```
