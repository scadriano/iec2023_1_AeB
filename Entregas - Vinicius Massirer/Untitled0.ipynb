```python
import pandas as pd

print('Biblioteca carregada com sucesso!')
```

```python
df_vinicius = pd.read_csv ('https://raw.githubusercontent.com/scadriano/iec2023_1_AeB/main/01.%20Conceitua%C3%A7%C3%A3o%20de%20Python/hospital.csv', sep=',', index_col=None)

print('Dados importados com sucesso!')
```

```python
display(df_vinicius.head())
```

```python
display(df_vinicius)
```

```python
df_vinicius = df_vinicius.drop(columns={'Estado','Time'})
```

```python
display(df_vinicius)
```

```python
df_vinicius.groupby('Diagnóstico').count().plot.pie(y = 'Internação')
```

```python
import matplotlib.pyplot as plt
plt.plot(df_vinicius['Nome'], df_vinicius['Idade'], marker='o', label='Idade')
plt.plot(df_vinicius['Nome'], df_vinicius['Temperatura'], marker='^', label='Temperatura')
plt.plot(df_vinicius['Nome'], df_vinicius['Peso'], marker='s', label='Peso')

plt.xlabel('Nome')
plt.ylabel('Valores')

plt.title('Idade, Temperatura e Peso')

plt.legend()
plt.show
```

```python
import seaborn as sns

sns.set_theme()

sns.scatterplot(x='Idade', y='Temperatura', hue='Nome', data=df_vinicius)

plt.xlabel('Idade')
plt.ylabel('Temperatura')

plt.show()
```
